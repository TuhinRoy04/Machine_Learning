# Predict whether a people face cardiac attack or not

```python
import pandas as pd
import sys
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
df=pd.read_csv("cardio_train.csv")
df["year"]=(df["age"]/365).round(0)
yr=df["year"]
df=df.drop(["id","age","year"],axis=1)
df.insert(2,"year",yr)
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.0125, random_state=20)
#Name
Name=input("Your name : ")
#Gender
g=input("Gender (M/F): ")
if(g=="M"):
    gen=2
elif(g=="F"):
    gen=1
else:
    print("Your entered gender is not correct !!!")
    sys.exit()
#Height
h=int(input("Height (in cm) : "))
#Age
ag=int(input("Age : "))
#Weight
w=int(input("Weight (in kg) : "))
#Systolic bp
s=int(input("BP (Systolic) : "))
#Diastolic bp
d=int(input("BP (Diastolic) : "))
#Cholesterol
c=input("Cholesterol level (N = Normal ; A = Above Normal ; H = High) : ")
if(c=="N"):
    chol=1
elif(c=="A"):
    chol=2
elif(c=="H"):
    chol=3
else:
    print("Your entered cholesterol level is not correct !!!")
    sys.exit()
#Glucose
sug=input("Sugar level (N = Normal ; A = Above Normal ; H = High) : ")
if(sug=="N"):
    sugar=1
elif(sug=="A"):
    sugar=2
elif(sug=="H"):
    sugar=3
else:
    print("Your entered sugar level is not corrected")
    sys.exit()
#Smoke
sm=input("You smoke (Yes/No) ?? : ")
if(sm=="Yes"):
    smoke=1
elif(sm=="No"):
    smoke=0
else:
    print("You haven't entered correct choice !!!")
    sys.exit()
#Alcohol
al=input("You take alcohol (Yes/No) ?? : ")
if(al=="Yes"):
    a=1
elif(al=="No"):
    a=0
else:
    print("You haven't entered correct choice !!!")
    sys.exit()
#Active
e=input("You do exercise (Yes/No) ?? : ")
if(e=="Yes"):
    act=1
elif(e=="No"):
    act=0
else:
    print("You haven't entered correct choice !!!")
    sys.exit()
print("\nYour result is now processing.....\n")
clf=RandomForestClassifier()
clf=clf.fit(X_train,y_train)
pred=clf.predict([[gen,h,ag,w,s,d,chol,sugar,smoke,a,act]])
if (pred==0):
    print("Result : No, you have not faced cardiac attack")
else:
    print("Result : Yes, you have faced cardiac attack")
accuracy=str((clf.score(X_test,y_test)*100).round(3))
percentage=str("%")
cat=accuracy+" "+percentage
print("\nYour dataset's accuracy score is :",cat)
```
